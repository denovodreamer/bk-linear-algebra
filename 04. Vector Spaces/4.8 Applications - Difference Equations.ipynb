```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def y_function(t):
    return np.cos(12 * np.pi * (t) / 4)

def z_function(k): 
    return 0.35 * y_function(k+2) + 0.5 * y_function(k+1) + 0.35 * y_function(k)

k = np.arange(0, 30, 0.01)

y = y_function(k)
z = z_function(k)

plt.plot(k, y, label = "y_k") 
plt.plot(k, z, label = "z_k") 
plt.legend() 
plt.show()
```
